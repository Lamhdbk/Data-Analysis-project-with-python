```python
university_towns=[]
with open('D:\Data Analytics Real World Project Python\Python data cleaning/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            # remember this state until the next is found
            state=line
        else:
            # otherwise, we have a city, keep state as las-seen
            university_towns.append((state, line))
```

```python
university_towns
```

```python
towns_df=pd.DataFrame(university_towns,columns=['State','RegionName'])
towns_df.head()
```

```python
def get_citystate(x):
    if ' (' in x:
        return x[:x.find(' (')]
    elif '[' in x:
            return x[:x.find('[')]
    else:
        return x
```

```python
# sau đó dùng applymap() theo hàm,

towns_df=towns_df.applymap(get_citystate)
```
